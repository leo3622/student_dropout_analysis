# Data Processing

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


from scripts.graph import plot_graph, save_graph

# Load the dataset
df = pd.read_csv('data/student_data.csv', encoding='utf-8', sep=';')
df.head()
```

```python
# Display basic information about the dataset
df.info()
df.describe()
```

```python
# Check for missing values
missing_values = df.isnull().sum()
print("Missing values in each column:")
print(missing_values)
```

```python
# View column names
print("Column names in the dataset:")
print(df.columns.tolist())
```

```python
# How student performance affects their dropout rate
df['performance_score'] = (df['Curricular units 1st sem (grade)'] + df['Curricular units 2nd sem (grade)']) / 2

# Calculate average performance score by output group
avg_score_each_group = df.groupby('Output')['performance_score'].mean()

avg_score_each_group = avg_score_each_group.reset_index()
avg_score_each_group.columns = ['Output Group', 'Average Performance Score']


graph = plot_graph(avg_score_each_group, 'Output Group', 'Average Performance Score', 'Average Performance Score by Output Group', 'Output Group', 'Average Performance Score', ['skyblue', 'orange', 'green'], 'bar', legend=False)
save_graph(graph, '../pictures/avg_performance_score_by_output.png')
```

```python
# How evening classes affect student output
num_to_category = {1: 'Day time', 0: 'Evening time'}
df['evening attendance'] = df['evening attendance'].map(num_to_category)
```

```python
# Count the number of students in each output group for evening and day time classes
output_counts = pd.crosstab(df['evening attendance'], df['Output'])
output_counts = output_counts.reset_index()
graph = plot_graph(output_counts, 'evening attendance', ['Dropout', 'Enrolled', 'Graduate'], 'Student Output by Class Time',
                   'Class Time', 'Number of Students', ['skyblue', 'orange', 'green'], 'bar', legend=True, stacked=True)
save_graph(graph, '../pictures/student_output_by_class_time.png')
```

```python
# Gender and output group analysis
gender_decoder = {1: 'Male', 0: 'Female'}
df['Gender'] = df['Gender'].map(gender_decoder)
```

```python
gender_output_counts = pd.crosstab(df['Gender'], df['Output'])
avg_dropout_by_gender = (
    gender_output_counts['Dropout'] /
    gender_output_counts.sum(axis=1) * 100
)
avg_dropout_by_gender.plot(kind='bar', color=['skyblue', 'orange'])
plt.title('Average Dropout Rate by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Dropout Rate (%)')
plt.xticks(rotation=0)
plt.savefig('../pictures/avg_dropout_by_gender.png')
plt.show()
```

```python
# Nationality distribution
nationality_decoder = {
    1: "Portuguese",
    2: "German",
    6: "Spanish",
    11: "Italian",
    13: "Dutch",
    14: "English",
    17: "Lithuanian",
    21: "Angolan",
    22: "Cape Verdean",
    24: "Guinean",
    25: "Mozambican",
    26: "Santomean",
    32: "Turkish",
    41: "Brazilian",
    62: "Romanian",
    100: "Moldova (Republic of)",
    101: "Mexican",
    103: "Ukrainian",
    105: "Russian",
    108: "Cuban",
    109: "Colombian"
}

df.rename(columns={'Nacionality': 'Nationality'}, inplace=True)
df['Nationality'] = df['Nationality'].map(nationality_decoder)
```

```python
nationality_count = df['Nationality'].value_counts().reset_index()
nationality_count.columns = ['nationality', 'count']
nationality_count
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='count', y='nationality', data=nationality_count, palette='pastel')
plt.xscale('log')
plt.title('Student Distribution by Nationality')
plt.xlabel('Number of Students')
plt.ylabel('Nationality')
plt.savefig('../pictures/student_distribution_by_nationality.png')
plt.show()
```
